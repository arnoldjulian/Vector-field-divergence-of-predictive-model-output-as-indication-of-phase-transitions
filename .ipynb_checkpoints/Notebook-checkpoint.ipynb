Vector field divergence of predictive model output as indication of phase transitions
-------------

-- Data from the Kuramoto Hopf model


```python
from utils import construct_partitions, roll, flip
from glob import glob
import numpy as np
import random
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
from torchvision import datasets, models, transforms
from train import train_and_evaluate_model

from cnn import Net3 # import the convolutional neural network

import matplotlib.pyplot as plt
%matplotlib inline

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
# create the training and test sets

partition={}

N_test = 5
N_train = 5

partition['test'],partition['train']=construct_partitions("data/*/",N_test, N_train)    


random.shuffle(partition['train'])
random.shuffle(partition['test'])

print('Total number of data samples: ', len(partition['test'])+len(partition['train']))
print('Total number of training samples: ', len(partition['train']))
print('Total number of validation samples: ', len(partition['test']))

N_grid = len(glob('data/*/'))
print('Number of grid points in parameter scan: ', N_grid)
```

```python
imSize = 128 # ImageSize of the Hopf oscillators
num_epochs = 80 # number of epochs that the model is trained

# initialize the convolutional neural network
model =  Net3()
model.to(device)

print(model)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0005)
# Decay LR by a factor of 0.5 every 20 epochs
exp_lr_scheduler = lr_scheduler.StepLR(optimizer, step_size=20, gamma=0.5)

# define data transformations for data augmentation
data_transforms = {
    'train': transforms.Compose([
        transforms.Lambda(lambda x: roll(x, 1, np.random.randint(imSize))),
        transforms.Lambda(lambda x: roll(x, 2, np.random.randint(imSize))),
        transforms.Lambda(lambda x: flip(x, 1 ,np.random.randint(2))),
        transforms.Lambda(lambda x: flip(x, 2 ,np.random.randint(2)))
    ]),
    'test': transforms.Compose([
        transforms.Lambda(lambda x: roll(x, 1, np.random.randint(imSize))),
        transforms.Lambda(lambda x: roll(x, 2, np.random.randint(imSize))),
        transforms.Lambda(lambda x: flip(x, 1 ,np.random.randint(2))),
        transforms.Lambda(lambda x: flip(x, 2 ,np.random.randint(2)))
    ])
}
```

```python
# train and evaluate the model in each epoch on training and test set, respectively.
_, _, _, divp, p0, deltap = train_and_evaluate_model(model, criterion, optimizer, exp_lr_scheduler, 
                                                        partition, data_transforms, imSize, num_epochs)
```

```python
# plot {delta_p, div p} vs p_0
epoch = num_epochs-1


# cut the first and last value as there are large peaks in the divergence due to the numerical computation
plt.plot(p0[1:-2],2*deltap[epoch][1:-2], label ='$2\delta p_0$')
plt.plot(p0[1:-2],divp[epoch][1:-2], label ='$\mathrm{div}(\delta p)$')
plt.xlabel('$p_0$')
plt.legend()

plt.show()
```

```python
# inspect a few samples

from torch.utils import data
from dataloader import Dataset
import torchvision


loaded_set = Dataset(partition['train'], imSize, data_transforms['train'])
loader = data.DataLoader(loaded_set, batch_size=4, shuffle=True, num_workers=1)


states, p0i = next(iter(loader))
          

for index in range(states.shape[0]):
    # 0 : cos(phi), 1 : sin(phi), 2: phidot
    plt.imshow(states[index,0,:,:], origin='lower', cmap='hsv')
    plt.title('S2 = {:f}'.format(np.exp(p0i[index].item())))
    plt.colorbar()
    plt.show()
```
